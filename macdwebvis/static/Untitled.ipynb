```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn as sb

df = pd.read_csv("menu.csv")
df_main = pd.concat([df["Category"],df["Calories"],df['Total Fat'], df["Cholesterol"], df["Sodium"], df["Carbohydrates"], df["Sugars"], df["Protein"], df["Vitamin A (% Daily Value)"], df["Vitamin C (% Daily Value)"]
],axis=1)
by_category = df_main.groupby("Category")

#each drink has 3 size(small, medium, large) 
by_category.size().plot(kind='bar')

zscore = lambda x: (x - x.mean()) / x.std()

by_category.plot(kind='box',figsize=(12,5))
mean = by_category.mean()

mean_norm = mean.transform(zscore) + 2
mean_norm.plot(kind='bar',figsize=(15,5))
```
